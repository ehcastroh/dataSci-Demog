## And what might the  logistic growth model have to do with the three humps  ?

Although it is both pretty and useful,  the logistic model is just a two parameter family of curves -- In other words, it is no more than a mathematical way of drawing s-shaped curves.  It contains no biological, ecological or demographic *mechanism* whereby some level of population density drives down the population growth rate. 

What we have in the logistic model is something that describes well things like asexually reproducing bacteria growing in a petri dish where all units of the population are reproductively equal, and the concept of carrying capacity is kind of sensible.  The story of the three humps is quite different. It is the story of how technology changed the trajectory of human population growth.  So in modern times, with technology not being what you'd call constant, human population trajectories have not behaved logistically. 

A somewhat famous example of this is from Raymond Pearl who in the 1930s fitted a logistic function to US population data since 1790 and confidently predicted that the US population would level off at 198 million.  Here is his data and graph courtesy of the Monkey's Uncle blog http://monkeysuncle.stanford.edu/?p=93

![Raymond Pearl's logistic estimate](http://courses.demog.berkeley.edu/mason88/images/pearl-badfit.png)


<b>Key idea: </b> For human populations about which we know a lot e.g. modern human populations, the logistic model has proved to be not very good.  However,  for  populations from the more distant past -- <i>populations that we don't know so much about</i>, the logistic model is somewhat more appealing.

```python
##  And The logistic growth model ... The first hump
##  

def dNdt(N,gamma=.04,K=10**6) :
    """
    for N,gamma and K  return logistic growth rate
    """
    return(gamma*N*(1-(N/K)))

K=10**6  # world population 10K BCE .. roughly
N0= 2 # starting population

gamma= ?  ## What max growth rate gets us to 1 million in 150K years ?
grates=[0]

while N0 < K - 1 :
    N1=N0+dNdt(N0,K=K,gamma=gamma)
    g=N1/N0 -1   
    #print("N0={0} N1={1} g={2}".format(N0,N1,g))
    grates.append(g)
    Npop.append(N1)
    N0=N1

poptab1=Table().with_columns('year',np.arange(len(Npop)),
                             'N',Npop,
                             'g',grates)
poptab1.plot(select=['N'],column_for_xticks='year')
poptab1.plot(select=['g'],column_for_xticks='year')
```

## swe1751  is called a 'cohort life table'

It contains everything demographers think they need to know about the mortality experience of the 'cohort' of Swedes who were born in 1751.  We're using Swedish data here because, unlike every other country on earth, the Swedes collected this data way back then.  Despite being a long time ago, 1751 is an interesting year because right about this time, the English are starting to mechanize cloth production and thereby launching ...the Industrial Revolution -- which is and was a very big deal both demographically and in every other way.

The structure of the life table reflects its ancient origins  (the first was published in 1662 by John Graunt) in that three of the columns are interesting and the rest are just intermediate calculations -- they are handy for various other obscure things that demographers do but mostly they are in the table so that a person with a modern computer can construct the $e_x$ column, described below.

### The column named 'lx' (lower case L)  is the starting point.  
It contains the number of people who survive to the beginning of each Age.  In the top row, Age =0 refers to birth -- the beginning of year of age 0.  The lx column shows 100000 -- How convenient, you might think, that such a round number of Swedes happened to be born in 1751-4.   But no -- All the numbers in the lx column have been multiplied by the $\frac{100,000}{actual number of births}$  Often one divides the lx column by that 1000000 which gives us the proportion of people born who survive to the beginning of each year of Age.

"Normalizing" the cohort size,  makes the lx curve easier to work with (important with the computing power available in 1661) -- also easier to compare across years -- as you see in the graphs in the next cell.

## The column in which most people are most interested is $e_x$ the expected years of remaining life at each age  aka  *life expectancy*.
The number in the first row of the $e_x$ column is the expected remaining years of life at birth -- or more commonly the "life expectancy". It requires quite a few calculations to derive $e_x$ from $l_x$ and when you minor in Demography you will learn all about these calculations in a way that will be far more interesting that what we can do here.

The important thing is to understand is what the $e_x$ means and to do that, we're going to compute it in a 21st century data sciency way.  -- By generating a simulated population with death ages corresponding to $lx$.  From our simulated population or list of life spans, we can compute all sorts of interesting values which characterize what life (and death) was like over the centuries.

```python
## adding a colum table to swe1751
def pdeathF (ltab):
    """
    Adds a column to a life table the probability of dying during each Age conditional only on being born.
    """
    ## np.diff(x) returns an array constructed by subtracting the nth element of x from the n+1 th element
    ## since lx is decreasing the result is negative but its the negative of the number of deaths that ocurred
    ## during each age interval
    ## take absolute value to flip the sign
    
    pdeath= abs(np.diff(ltab['lx'])/ltab['lx'][0])
       
    # generally the prob of surviving top age is zero but it doesn't have to be
    ltab.append_column('pdeath',np.append(pdeath,1-pdeath.sum()))
    return(ltab)
    
```

```python
## Using the pdeathF  function ...
swe1751=pdeathF(swe1751)
swe1751


##note that pdeath and qx are the same in row Age 0  but not so much thereafter
```

## How is it possible to compute a life expectency if no one in the cohort has reached the age of 2 yet?

Good question!

The answer is at once complicated and yet simple. Life tables begin with the lx column. In the case of cohort, this is quite natural because one starts with a certain number of people born in a certain year and then we just watch them die -- well we really count birth and death records,  the entire cohort has to be pretty much extinct in order for us to construct a life table.  

What we do for a cohort of say 1 year olds who were born last year is different of course.  For them we create what is called a *synthetic cohort*  we do it by watching the entire population for (generally) one year.  After a year we count up all of the deaths at each age and from that we can compute the $q_x$ column of the synthetic cohort life table.  $q_x$ are the probability of death within the age category conditional on surviving to the beginning of the age. From the $q_x$ column we can compute an lx by simply starting with a big round number like 100,000 calling it $l_0$ and multiplying it by $q_0$ to get $l_1$. Then we multiply $l_1$ by $q_1$ to get $l_2$ and so on.

The details are not too important as long as you can convince yourself that it is possible to go from an array of age specific probabilities of death to $l_x$. 

But what exactly is that synthetic lx that we just computed? It is the number of people who *would* survive to each age from a cohort of people whose mortality experience at each age matched the mortality experience of each age group in the population over a particular year.

Once we have a synthetic $l_x$  we can do all the same life table exercises as with the cohort data.

```python
# a function to generate a simulated population
def popSim(ltab,N=100000):
    """
    uses deathF to add a column of probabilities of death; then draws N ages (at death) 
    """
    ltab0=pdeathF(ltab)
    ## zero probability of death is fine for ages to which no one survives -- but
    ## .sample() cannot stand it so we remove them from consideration
    ltab0=ltab0.where('pdeath',are.above(0))
    aged=ltab0.sample(k=N,with_replacement= True, weights= ltab0['pdeath'])['Age']
    return(aged)
```
