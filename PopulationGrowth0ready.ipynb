## Demography starts with the most fundamental of fundamental equations

$$Populaion_t=Population_{t-1}+Births_{t-1}-Deaths_{t-1}+Immigration_{t-1} -Emigration{t-1}$$


### Since we are discussing world population we can simplify this to

$$Populaion_t=Population_{t-1}+Births_{t-1}-Deaths_{t-1}$$

### To the extent the $Births_{t-1}$ is a function of $Population_{t-1}$  and $Deaths_{t-1}$ is a function of $Population_{t-1}$  we can simplify (or obfuscate) this to

$$P_{t}=f(P_{t-1},?)$$

Where P stands for Population; f() stands for some unspecified 'function' and '?' stands for some unknown function input (or 'parameter' or 'argument').  The important thing here is that because people and organisms cause their own reproduction, the future population is a function of the current population (and perhaps a little additional information)

One of a few things that distinguishes humans from this sort of plant is that we reproduce and die throughout the year.  The *exponential* growth model  handles that difference -- and has some very nice mathematical properties that you will learn about when you minor in demography.

## Part 2: Logistic growth

Geometric (and exponential) growth are one parameter models -- and that's nice, but the lesson so far is that these single paramter models do not describe the three humps of human history particularly well.  A closer fit -- at least for the first two humps can be achieved with a logistic model wherein the growth rate is dependent not only on the current population but also on the "carrying capacity" , K, and the assumed highest possible rate of growth which we'll call $\gamma$ (gamma).

$$ \frac{dN}{dt}=\gamma P \big(1-\frac{N}{K}\big)$$

## Don't panick.

While the equation above is techincally a differential equation... to use it we don't need to solve it -- we only need to *decode* the math and then *code* it in python. The left hand side, $\frac{dN}{dt}$ translates to "the change in N over the change in t"  where N is the number of organisms or the population, and t is time.  In other words, $\frac{dN}{dt}$ is the amount by which the population N grows in a unit of time t. For our purposes $dt=1$.  So we are left with a fairly simple idea that the change in N in one time period = $\gamma$  (which is just a number which in the case of humans is about .04) multiplied by the expression: $N(1-\frac{N}{K})$ which changes with N, but is not hard to compute. The carrying capacity "K", it a bit problematic, for reasons that hopefully we discussed in class. But for purposes of the logistic model, it's a just a number that the investigator has a strong belief about.

Take a moment to ponder how the expression on the right side of the equal sign changes as N goes from something close to zero to something close to K (which is presumably a large number).


(those who have taken calculus will notice that this is a bit of an oversimplification. But just deal with it OK).

## Let's code the logistic growth model 

While conceptually more complicated than the constant growth rate models that we have been working with, in python, there's not much difference.

We introduce in the next cell, the idea of a "function"  this has probably not been covered in the foundation course yet, but it will be soon.

```python
## dNdt below is a python "function"  it expects you to tell it what N is and it will tell you by how much
## N should grow in the coming single year period. If you want to you can also specify gamma and K but if you don't
## the function will use its default values .04 and a million
def dNdt(N,gamma=.04,K=10**6) :
    """
    for N,gamma and K  return logistic growth rate
    """
    return(gamma*N*(1-(N/K)))
```

The growth model that we have been using is called "geometric" growth.  It's a perfectly fine description of population growth in that, population in time t+1  is a function of population t plus some parameter.

But is it a particularly good model of Human Population Growth ?

The answer is yes for most purposes .. but not all

Suppose the initial population is 1000 and the growth rate is .1 (10%)

What is the mid year population ... at time t=0.5 ?

You could say that its 1000 + 1000*(.1/2)  or 1050.  That's seems reasonable, but it raises a question:

## if the midyear population is 1050 -- what is the growth rate during the second half of the year?



```python
ipop=1000
grate=.1
midyear_pop=ipop + ipop*(grate/2)

endyear_pop=1050+ 1050*(grate/2)

out="""
initial population ={0}
growth rate = {1}
midyear population estimate?={2}
midyear population after an additional half year of growth at {1} ={3}
""".format(ipop,grate,midyear_pop,endyear_pop)
print(out)
```


## Discuss what's going on here

One way to think of this is to consider a bank account -- where the rules are made by people rather than by nature. 
When a bank advertises an interest rate, it needs to specify the period of compounding. In the above example one would say that the annual interest rate is 10% compounded semi annually (every 6 months).  The more frequently the interest is "compounded"  the more frequently the accrued interest becomes part of the balance on which interest is earned.  So at any given interest rate,  the more frequently that compounding ocurs, the more the account owner will have at the end of the year -- up to ... as it turns out ... a limit.

For compounding semi annually we can write:

$$ P_1= P_0*\big(1+\frac{g}{2}\big)*\big(1+\frac{g}{2}\big) $$

where  g  is the growth rate 

if we want to compound more frequently than twice a year,  let's n times per year we could write:

$$P_1 = P_0\Big(1+\frac{g}{n}\Big)^n$$

With python we can  graph what's going on here

We're going to use a function - which is something that you have not yet seen so just kind of go with it.
a function is a chunk of stored code -- in this case this code take a growth rate, g and a number of compounding periods n  and outputs a table where one column, t holds the time and another column N, holds the population/bank balance at the corresponding time *IF* the balance were growing at annual rate g compounded n times per year.

Just run the cell and look at the output to see what's going on.


```python
# create a function called cpound
def cpound(g,n) :
    """
    outputs a list of n  factors by which P_0 would be multiplied to get intermediate values of P_0 which
    is growing at rate g per year compounded n times
    """
    res=[1]
    for i in np.arange(start=0,stop=n,step=1):
        res.append(res[i]*(1+g/n))
    Res=Table().with_column('P',res)
    Res.append_column('t',np.append(np.arange(start=0,stop=1,step=1/n),[1]))
    return(Res)

#execute the function cpound for g=.1 and n - 3
print('behold')
cpound(.1,3)
```

### As the number of compounding periods gets larger,  the value of $P_1$ rapidly approaches $P_0e^g$

### And for any time t  $ P_t=P_0e^{tg}$

# Populations that can be described by the equation $P_t=P_0e^{rt}$  are said to be growing "exponentially"


Notice that I just switched from using "g" to quantify the growth rate to "r". I did that to make the point that there are two distinct ways of quantifying growth rates.  Geometric growth and exponential growth. Demographers prefer to think in terms of exponential growth for human populations because (moderately large) human populations reproduce (and die) on a continuous time scale -- and also because it has an nice mathematical properties that you will learn more about after you minor in demography. 

For our purposes today -- in discussing the long term growth of human populations, it doesn't matter much which model we use -- the periods of early human existence we are generally interested in centuries or milenia rather than days or months if for no other reason, we have very little data to work with.
